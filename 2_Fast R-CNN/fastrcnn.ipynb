```python
test_path = '/Users/Hyungdanny/Desktop/Ai/CP2/DeeFashion2/test_data/image/000005.jpg'
```

```python
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
```

```python
# image를 불러 올 때 df에 있는 id 를 이용해야한다.
# 모든 데이터의 기준은 csv_df 에 있는 것으로 한다.

images = []
image = keras.utils.load_img(test_path, target_size=(224, 224),interpolation="nearest")# VGG 16 모델을 사용할 예정이기 때문에.
image = tf.keras.preprocessing.image.img_to_array(image)
ss.setBaseImage(image)
ss.switchToSelectiveSearchFast()
ssresults = ss.process()
imout = image.copy()

# 같은 데이터에 있는 이미지를 불러 와야 한다. 한 이미지에 아이템이 2개가 있는 경우에는 두번 들어간다.
# csv_df 에 있는 bounding box 값을 가져온다.
for i in range(len(csv_df)):
    x = csv_df['bounding_box'][i][0]
    y = csv_df['bounding_box'][i][1]
    w = csv_df['bounding_box'][i][2]
    h = csv_df['bounding_box'][i][3]

# selectiveSearch로 얻어진 값을 가지고 온다.
    for e,result in enumerate(ssresults):
        if e < 2000:
            sx,sy,sw,sh = result[e]
            iou = tf.keras.metrics.MeanIoU(num_classes=13) # keras library로 iou 값을 계산.
            iouValue = iou.update_state([x, y, w, h], [sx, sy, sw, sh]) # selectivesearch로 얻은 값으로 iou계산.
                 
            if iouValue > 0.70:
                        
                simage = keras.utils.load_img(imout[y:y+h,x:x+w], target_size=(224, 224),interpolation="nearest")# VGG 16 모델을 사용할 예정이기 때문에.
  
                images.append(simage)
                                
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
```

```python
vgg16_base = tf.keras.applications.VGG16(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=12,
    classifier_activation="softmax",
)
```

```python
model = Sequential()
model.add(Conv2D(input_shape=(224,224,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Flatten())
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=2, activation="softmax"))
```
